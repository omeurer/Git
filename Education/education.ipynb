Les données sont axées sur le baccalauréat.

```python
data.columns
```

```python
dfColNeeded=data.iloc[:,1:] #On enlève le type d'établissement
```

```python
dfColNeeded
```

On cherche à sommer le nombre d'élèves selon deux catégories : ils ont passé le bac en avance et/ou à l'heure, ou en retard. 

```python
dfColNeeded.iloc[:,1:3] #vérification pour somme avance ou à l'heure
```

```python
dfColNeeded.iloc[:,3:5] #verification bac retard ou non spécifié
```

Nombre d'élèves ayant passé le bac :

```python
dfColNeeded.iloc[:,1:5]
```

On somme pour connaitre le nombre d'élèves totaux. 

```python
effectifBac=dfColNeeded.iloc[:,1:5].sum(axis=1)
```

```python
effectifBac
```

On calcule la proportion de bacheliers en avance:

```python
dfColNeeded['AgeBacAvanceOuATemps']=dfColNeeded.iloc[:,1:3].sum(axis=1)
```

```python
dfColNeeded['AgeBacAvanceOuATemps']=dfColNeeded['AgeBacAvanceOuATemps']/effectifBac
```

```python
dfColNeeded
```

On calcule la proportion de bacheliers en retard:

```python
dfColNeeded['AgeBacRetard']=dfColNeeded.iloc[:,3:5].sum(axis=1)
```

```python
dfColNeeded['AgeBacRetard']=dfColNeeded['AgeBacRetard']/effectifBac
```

```python
dfColNeeded
```
