```python
def my_geocoder(row):
    try:
        point = geocode(row, provider='nominatim',user_agent='autogis_xx').geometry.iloc[0]
        return pd.Series({'Latitude': point.y, 'Longitude': point.x, 'geometry': point})
    except:
        return None
```

```python
df=join_test_f
column_name='LIBCOM'
```

```python
df[['Latitude', 'Longitude', 'geometry']] = df.apply(lambda x: my_geocoder(x[column_name]), axis=1)

print("{}% of addresses were geocoded!".format(
    (1 - sum(np.isnan(df["Latitude"])) / len(df)) * 100))

# Drop df that were not successfully geocoded
df = df.loc[~np.isnan(df["Latitude"])]
df = gpd.GeoDataFrame(df, geometry=df.geometry)
df.crs = {'init': 'epsg:4326'}
df
```

```python
# Create a base map
m_5 = folium.Map(location=[54,15], tiles='cartodbpositron', zoom_start=4)

# Add a heatmap to the base map
HeatMap(data=df[['Latitude', 'Longitude','DISP_TP6018']], radius=10).add_to(m_5)

# Display the map 
m_5
```
