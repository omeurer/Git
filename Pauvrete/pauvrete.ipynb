Pour faire la jointure entre les données relatives à la pauvreté et le dataset de références iris, il faut ajouter un 0 aux codes iris n'en ayant pas (comme pour l'Ain 01).

```python
for index, iris in zip(data.index, data.IRIS):
    if len(iris)!=9:
        data.loc[index,'IRIS'] = data.loc[index,'IRIS'].zfill(9)
```

#### Rename ref iris for join column

#### Join columns

```python
irisDepCom
```

```python
pauvAvecDep = data.merge(irisDepCom, how='left',on='IRIS')
```

#### Groupy and mean par département

```python
dfGroupByDep=pauvAvecDep.groupby(by=['numDep']).mean()
```
