On ne garde les données que post 2010. 

```python
dfPost2010=data[data['Date début']>'2010-01-01'].drop(labels=['INSEE','Date fin'],axis=1)
```

```python
dfPost2010['Périls'].value_counts()
```

Nous allons donc regrouper les évènements dont les dates sont postérieures à 2010, et pour les types de catastrophe les plus fréquentes.

```python
typesInondations=data['Périls'].value_counts()
```

```python
dfPost2010
```

Types de catégories : 
* inondations, mouvement et effondrement de terrain,
* catatrophes liées aux actions des vagues, 
* séisme, 
* avalanches. 

```python
listeInondations=['Inondations et coulées de boue',
                'Mouvements de terrain différentiels consécutifs à la sécheresse et à la réhydratation des sols',
                'Inondations, coulées de boue et mouvements de terrain',
                'Mouvements de terrain',
                'Inondations par remontées de nappe phréatique']

listeVagues=["Inondations, coulées de boue, mouvements de terrain et chocs mécaniques liés à l'action des vagues",
"Inondations et chocs mécaniques liés à l'action des vagues",
"Chocs mécaniques liés à l'action des vagues",
"Inondations, coulées de boue et chocs mécaniques liés à l'action des vagues"]

listeAvalanches=['Avalanche']

listeSeisme=["Séisme"]                                                                                                   
```

```python
dfpost2010Dummies=dfPost2010.copy()
```

```python
dfpost2010Dummies["Périls"].isin(listeInondations).astype(int)
```

On convertit en dummies pour faciliter le calcul sur data studio:

```python
dfpost2010Dummies["Inondations, mouvement et effondrement de terrain"]=dfpost2010Dummies["Périls"].isin(listeInondations).astype(int)
dfpost2010Dummies["Inondations liées à l'action des vagues"]=dfpost2010Dummies["Périls"].isin(listeVagues).astype(int)
dfpost2010Dummies["Avalanches"]=dfpost2010Dummies["Périls"].isin(listeAvalanches).astype(int)
dfpost2010Dummies["Seismes"]=dfpost2010Dummies["Périls"].isin(listeSeisme).astype(int)
```

On ne garde que l'année en question:

```python
dfpost2010Dummies['Annee'] = pd.DatetimeIndex(dfpost2010Dummies['Date début']).year
dfpost2010Dummies=dfpost2010Dummies.drop(columns='Date début')
```

```python
dfpost2010Dummies
```

On remplace les départements de la Corse par les numéro 291 et 292 pour faciliter les jointures par la suite (conflit avec le type de données entre "numéro de département" et "2A, 2B" pour la Corse).

```python
dfpost2010Dummies=dfpost2010Dummies.replace('2A',291)
dfpost2010Dummies=dfpost2010Dummies.replace('2B',292)
```

```python
dfpost2010Dummies
```

```python
dfpost2010Dummies=dfpost2010Dummies.drop(columns="Périls")
```

On somme par année et par département:

```python
dfDataStudio=dfpost2010Dummies.groupby(by=['Annee','Departement']).sum()
```
